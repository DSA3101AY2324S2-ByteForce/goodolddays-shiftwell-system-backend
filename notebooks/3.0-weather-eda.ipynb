### Data Preprocessing and EDA on Weather Data

#### 1. Data CLeaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df_weather_raw = pd.read_csv("../data/raw/weather_data.csv")
df_weather_raw.head()
```

```python
df_weather_raw.dtypes
```

```python
df_weather_raw.replace('-', np.nan, inplace=True)
```

```python
df_weather_raw.drop('Station', axis=1, inplace=True)
for col in df_weather_raw.columns:
    if df_weather_raw[col].dtype == 'object':
        df_weather_raw[col] = df_weather_raw[col].astype(float)
df_weather_raw['Date'] = pd.to_datetime(df_weather_raw[['Year', 'Month', 'Day']])
df_weather_raw['Date'] = pd.to_datetime(df_weather_raw['Date'], format='%Y-%m-%d')

df_weather_raw.head()
```

```python
df_weather_raw.isnull().sum()
```

There are some missing values in the weather data. We will fill the missing values with the median of that month in the corresponding column.

```python
# fill in na values with the median of the month 
df_weather = df_weather_raw.copy()
df_weather.fillna(df_weather_raw.groupby('Month').transform('median'), inplace=True)
df_weather.isnull().sum()
```

```python
df_weather.to_csv('../data/processed/weather_data_cleaned.csv', index=False) 
```
